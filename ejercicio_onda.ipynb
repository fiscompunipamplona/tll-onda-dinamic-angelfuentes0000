```python
from pylab import plot, imshow, show, xlabel, ylabel, pink
from numpy import loadtxt, array, zeros, sqrt, pi, sin, linspace
import numpy as np
from IPython import display as disp
import matplotlib.pyplot as plt
salida2=open("Ondas_Est.dat4", "w")
gral=800
glob=100
t=100
w=linspace(0,glob,gral)
z=linspace(0,glob,gral)
X,Y=np.meshgrid(w,z)
dat=np.zeros((gral,gral,t),dtype=float)
lda = 5.0 # cm
lda1 = 4.0
k = 2.*pi / lda
k1 = 2.*pi / lda1
zeta0 = 3.0 # cm
x0 = 75. # cm
y0 = 50. # cm
x1 = 25.
y1 = 50.
W = 100
for t in range(0,t):
        r=sqrt((X-x0)**2 + (Y-y0)**2)
        e=zeta0*sin(k*r+W*t)
        
        r1=sqrt((X-x1)**2 + (Y-y1)**2)
        e1=zeta0*sin(k1*r1+W*t)
        
        e21=e+e1
        #e21=e
        plt.gca().cla()
        dat[:,:,t]=e21
        #imshow(e21, origin="lower",extent=[0,glob,0,glob])
        imshow(e21, origin="lower", extent=[0, 100, 0, 100], aspect=1.0) # conserva la misma configuración 
        pink()
        xlabel("eje x cm")
        ylabel("eje y cm ")
        disp.clear_output(wait=True)
        disp.display(plt.gcf())
```

```python
for r in range(t):
    salida2.write('#Matriz %d \n' %(r))
    for i in range(gral):
        for j in range(gral):
            salida2.write('%0.4f ' %(dat[i,j,r]))
        salida2.write('\n')
salida2.close()
```
